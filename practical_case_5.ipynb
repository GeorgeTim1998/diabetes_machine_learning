```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import pandas as pd
from sklearn.decomposition import PCA
from mlxtend.plotting import plot_decision_regions
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/aridata/diabetes.csv')
df.head()
```

```python
X2 = df.drop(['Outcome'], axis=1)
y = df.Outcome
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.3, random_state=0)
```

```python
# Обучение классификатора
pca = PCA(n_components=2)
X = pca.fit_transform(X_train)
svm = SVC(C=1, kernel='linear')
svm.fit(X, y_train)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X, y_train)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y_train)
```

```python
# Отрисовка
plot_decision_regions(X, y_train.values, clf=svm, legend=2)
plt.show()
```

```python
plt.figure(figsize=(24,24))
plot_tree(clf, max_depth=3)
plt.show()
```

```python
plot_decision_regions(X, y_train.values, clf=knn, legend=2)
plt.show()
```
